```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("/kaggle/input/bank-customer-churn-modeling/Churn_Modelling.csv")
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
x = df[['CreditScore','Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]

y = df['Exited']
```

```python
sns.countplot(x=y);
```

```python
y.value_counts()
```

```python
# used for sampling and making both the y counts 1 and 0 same after getting low recall,
from imblearn.over_sampling import RandomOverSampler
```

```python
model = RandomOverSampler(random_state=0)
x_res, y_res = model.fit_resample(x ,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_res)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y_res,random_state=0,test_size=0.25)
```

```python
x.shape
```

```python
x_train.shape
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(100,100,100),max_iter=100, random_state=0)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay , classification_report, accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#getting good recall in 1 after getting oversampling
```

```python
accuracy_score(y_test,y_pred)
```
