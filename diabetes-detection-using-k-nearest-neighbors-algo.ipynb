Implementing K-Nearest Neighbors algorithm on diabetes dataset. Computing confusion
matrix, accuracy, error rate, precision and recall on the given dataset.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/diabetes/diabetes.csv")
df
```

```python
df.columns
```

```python
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'Pedigree', 'Age']]

y = df['Outcome']
```

```python
sns.countplot(x = y)
```

```python
y.value_counts()
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state = 0, test_size = 0.25)
```

```python
x.shape
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
tn, fp, fn, tp
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
error = []
for k in range(1,41):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    error.append(np.mean(pred != y_test))
```

```python
error
```

```python
plt.figure(figsize=(16,9))
plt.xlabel("value of k")
plt.ylabel("errors")
plt.grid()
plt.xticks(range(1,41))
plt.plot(range(1,41),error, marker='.')
```

```python
#the k value 33 has lowest errors,so,
model = KNeighborsClassifier(n_neighbors=33)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```
