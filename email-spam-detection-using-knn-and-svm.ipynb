To Classify the email using the binary classification method. Email Spam detection has two
states: a) Normal State – Not Spam, b) Abnormal State – Spam. Use K-Nearest Neighbors and
Support Vector Machine for classification. Analyze their performance.

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/email-spam-classification-dataset-csv/emails.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
x = df.drop(['Email No.','Prediction'],axis = 1)

y = df['Prediction']
```

```python
#always check that train n test data should be in int
x.dtypes
```

```python
import seaborn as sns
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state=0,test_size = 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
error = []
for k in range(1,41):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    error.append(np.mean(pred != y_test))
```

```python
error
```

```python
plt.figure(figsize=(16,9))
plt.xticks(range(1,41))
plt.grid()
plt.xlabel("value of predictions")
plt.ylabel("value of error")
plt.plot(range(1,41),error, marker='.')
```

```python
#having lowest error at value 1 so,
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
